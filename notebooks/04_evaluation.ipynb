```python
import sys
sys.path.append(r"C:\Users\Akash\Desktop\stock-forecasting\src")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.express as px

from data_loader import load_all
from prophet_model import (
    prepare_prophet_data, train_test_split_prophet,
    build_and_train_prophet, forecast_and_evaluate
)
from lstm_model import (
    prepare_lstm_data, split_and_scale,
    build_lstm_model, train_lstm, evaluate_lstm
)

print("✓ All imports successful")
```

```python
TICKER   = "AAPL"
LOOKBACK = 60

data_raw = load_all(save=False)
df       = data_raw[TICKER]

# --- Prophet ---
prophet_df          = prepare_prophet_data(df)
train_p, test_p     = train_test_split_prophet(prophet_df, test_ratio=0.2)
prophet_model       = build_and_train_prophet(train_p)
_, test_fc, p_metrics = forecast_and_evaluate(prophet_model, train_p, test_p)

# --- LSTM ---
close_df = prepare_lstm_data(df, TICKER)
X_train, y_train, X_test, y_test, scaler, split_idx = split_and_scale(close_df, lookback=LOOKBACK)
lstm_model = build_lstm_model(lookback=LOOKBACK)
train_lstm(lstm_model, X_train, y_train, epochs=50, batch_size=32)
predictions, actual, l_metrics = evaluate_lstm(lstm_model, X_test, y_test, scaler)

print("\n✓ Both models trained")
print(f"\nProphet → MAE: {p_metrics['MAE']} | RMSE: {p_metrics['RMSE']} | MAPE: {p_metrics['MAPE']}%")
print(f"LSTM    → MAE: {l_metrics['MAE']} | RMSE: {l_metrics['RMSE']} | MAPE: {l_metrics['MAPE']}%")
```

```python
test_dates = close_df.index[split_idx:]
train_dates = close_df.index[:split_idx]

fig = make_subplots(
    rows=1, cols=2,
    subplot_titles=["Prophet Forecast vs Actual", "LSTM Forecast vs Actual"],
    shared_yaxes=True
)

# Prophet
fig.add_trace(go.Scatter(x=test_p["ds"], y=test_p["y"],
    name="Actual", line=dict(color="#2196F3", width=2)), row=1, col=1)
fig.add_trace(go.Scatter(x=test_fc["ds"], y=test_fc["yhat"],
    name="Prophet", line=dict(color="#FF9800", width=2)), row=1, col=1)
fig.add_trace(go.Scatter(
    x=pd.concat([test_fc["ds"], test_fc["ds"][::-1]]),
    y=pd.concat([test_fc["yhat_upper"], test_fc["yhat_lower"][::-1]]),
    fill="toself", fillcolor="rgba(255,152,0,0.12)",
    line=dict(color="rgba(0,0,0,0)"), name="95% CI", showlegend=True
), row=1, col=1)

# LSTM
fig.add_trace(go.Scatter(x=test_dates, y=actual.flatten(),
    name="Actual", line=dict(color="#2196F3", width=2), showlegend=False), row=1, col=2)
fig.add_trace(go.Scatter(x=test_dates, y=predictions.flatten(),
    name="LSTM", line=dict(color="#E91E63", width=2)), row=1, col=2)

fig.update_layout(
    title=f"{TICKER} — Prophet vs LSTM Side-by-Side",
    height=450, template="plotly_white",
    legend=dict(orientation="h", yanchor="bottom", y=1.05)
)
fig.show()
```

```python
metrics_df = pd.DataFrame({
    "Model"  : ["Prophet", "LSTM"],
    "MAE"    : [p_metrics["MAE"],  l_metrics["MAE"]],
    "RMSE"   : [p_metrics["RMSE"], l_metrics["RMSE"]],
    "MAPE(%)"  : [p_metrics["MAPE"], l_metrics["MAPE"]]
})

fig = make_subplots(rows=1, cols=3,
    subplot_titles=["MAE (lower = better)",
                    "RMSE (lower = better)",
                    "MAPE % (lower = better)"])

colors = {"Prophet": "#FF9800", "LSTM": "#E91E63"}

for i, metric in enumerate(["MAE", "RMSE", "MAPE(%)"]):
    for _, row in metrics_df.iterrows():
        fig.add_trace(go.Bar(
            x=[row["Model"]], y=[row[metric]],
            name=row["Model"],
            marker_color=colors[row["Model"]],
            showlegend=(i == 0),
            text=[f"{row[metric]:.2f}"],
            textposition="outside"
        ), row=1, col=i+1)

fig.update_layout(
    title=f"{TICKER} — Prophet vs LSTM Metric Comparison",
    height=420, template="plotly_white",
    barmode="group"
)
fig.show()
```

```python
# Align lengths
prophet_residuals = test_p["y"].values - test_fc["yhat"].values
lstm_residuals    = actual.flatten() - predictions.flatten()

# Use shorter length if mismatch
min_len = min(len(prophet_residuals), len(lstm_residuals))
prophet_residuals = prophet_residuals[:min_len]
lstm_residuals    = lstm_residuals[:min_len]
dates_common      = test_dates[:min_len]

fig = make_subplots(rows=2, cols=2,
    subplot_titles=[
        "Prophet Residuals over Time", "LSTM Residuals over Time",
        "Prophet Residuals Distribution", "LSTM Residuals Distribution"
    ])

# Prophet residuals over time
fig.add_trace(go.Scatter(x=dates_common, y=prophet_residuals,
    line=dict(color="#FF9800", width=1), name="Prophet Residuals"), row=1, col=1)
fig.add_hline(y=0, line_dash="dash", line_color="black", row=1, col=1)

# LSTM residuals over time
fig.add_trace(go.Scatter(x=dates_common, y=lstm_residuals,
    line=dict(color="#E91E63", width=1), name="LSTM Residuals"), row=1, col=2)
fig.add_hline(y=0, line_dash="dash", line_color="black", row=1, col=2)

# Distributions
fig.add_trace(go.Histogram(x=prophet_residuals, nbinsx=40,
    marker_color="#FF9800", name="Prophet Dist"), row=2, col=1)
fig.add_trace(go.Histogram(x=lstm_residuals, nbinsx=40,
    marker_color="#E91E63", name="LSTM Dist"), row=2, col=2)

fig.update_layout(
    title=f"{TICKER} — Residuals Analysis",
    height=600, template="plotly_white", showlegend=False
)
fig.show()

print(f"\nProphet → Mean residual: ${prophet_residuals.mean():.3f} | Std: ${prophet_residuals.std():.3f}")
print(f"LSTM    → Mean residual: ${lstm_residuals.mean():.3f} | Std: ${lstm_residuals.std():.3f}")
```

```python
def rolling_mape(actual, predicted, window=20):
    errors = np.abs((actual - predicted) / actual) * 100
    return pd.Series(errors).rolling(window).mean().values

p_rolling = rolling_mape(test_p["y"].values[:min_len], test_fc["yhat"].values[:min_len])
l_rolling = rolling_mape(actual.flatten()[:min_len], predictions.flatten()[:min_len])

fig = go.Figure()
fig.add_trace(go.Scatter(x=dates_common, y=p_rolling,
    name="Prophet Rolling MAPE", line=dict(color="#FF9800", width=2)))
fig.add_trace(go.Scatter(x=dates_common, y=l_rolling,
    name="LSTM Rolling MAPE", line=dict(color="#E91E63", width=2)))

fig.update_layout(
    title=f"{TICKER} — Rolling 20-Day MAPE over Time",
    xaxis_title="Date", yaxis_title="MAPE (%)",
    height=420, template="plotly_white"
)
fig.show()
```

```python
prophet_csv = pd.DataFrame({
    "Ticker": [TICKER],
    "MAE": [p_metrics["MAE"]],
    "RMSE": [p_metrics["RMSE"]],
    "MAPE": [p_metrics["MAPE"]]
}).set_index("Ticker")

prophet_csv.to_csv("data/prophet_metrics.csv")
lstm_csv    = pd.read_csv("data/lstm_metrics.csv",    index_col="Ticker")

prophet_csv.columns = [f"Prophet_{c}" for c in prophet_csv.columns]
lstm_csv.columns    = [f"LSTM_{c}"    for c in lstm_csv.columns]

final = pd.concat([prophet_csv, lstm_csv], axis=1)
final = final[["Prophet_MAE","LSTM_MAE",
               "Prophet_RMSE","LSTM_RMSE",
               "Prophet_MAPE","LSTM_MAPE"]]

# Add winner column
final["Winner"] = final.apply(
    lambda r: "LSTM ✓" if r["LSTM_MAPE"] < r["Prophet_MAPE"] else "Prophet ✓", axis=1
)

print("\n========== FINAL MODEL COMPARISON ==========")
print(final.to_string())
final.to_csv("data/final_comparison.csv")
print("\n✓ Saved to data/final_comparison.csv")
```

```python
limitations = """
## Limitations & Honest Analysis

1. **Look-ahead bias**: Models trained on split data; no future data leaked.
2. **No fundamental data**: We use only price/volume — earnings, news, macros ignored.
3. **LSTM overfits on low-volatility periods**: Rolling MAPE spikes during high
   volatility (COVID crash, rate hike cycles) — model struggles with black swans.
4. **Prophet assumes additive seasonality**: Works well for trends but misses
   sharp discontinuities in price action.
5. **MAPE limitation**: Misleading when actual prices are near zero (not an issue
   for stocks, but worth noting for penny stocks).
6. **Prediction ≠ trading signal**: Lower MAPE does not mean profitable trading.
   This is a forecasting exercise, not a trading strategy.
"""
print(limitations)

# ensure we write using UTF-8 to avoid encoding errors on Windows
with open("data/limitations.md", "w", encoding="utf-8") as f:
    f.write(limitations)
print("✓ Saved to data/limitations.md")
```
